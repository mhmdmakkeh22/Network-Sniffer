```python
import socket
import struct
import textwrap
import time
```

```python
def main():
    conn = socket.socket(socket.AF_INET, socket.SOCK_RAW, socket.IPPROTO_IP)
    conn.bind(("192.168.1.101", 0))
    conn.setsockopt(socket.IPPROTO_IP, socket.IP_HDRINCL, 1)
    conn.ioctl(socket.SIO_RCVALL, socket.RCVALL_ON)

    start_time = time.time()
    total_bytes = 0

    while True:
        raw_data, addr = conn.recvfrom(65536)
        dest_mac, src_mac, eth_proto, data = ethernet_frame(raw_data)
        total_bytes += len(raw_data)

        elapsed_time = time.time() - start_time
        if elapsed_time > 1:  # Calculate bandwidth every second
            bandwidth = total_bytes / elapsed_time
            print(f"Bandwidth used: {bandwidth:.2f} bytes/second")
            start_time = time.time()
            total_bytes = 0

        print('\nEthernet Frame:')
        print('Destination MAC: {}, Source MAC: {}, Protocol: {}'.format(dest_mac, src_mac, eth_proto))
```

```python
def ethernet_frame(data):
    dest_mac, src_mac, proto = struct.unpack('! 6s 6s H', data[:14])
    return get_mac_addr(dest_mac), get_mac_addr(src_mac), socket.htons(proto), data[14:]
```

```python
def get_mac_addr(bytes_addr):
    bytes_str = map('{:02x}'.format, bytes_addr)
    return ':'.join(bytes_str).upper()

```

```python
if __name__ == "__main__":
    main()
```
